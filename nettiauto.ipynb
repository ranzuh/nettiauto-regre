```python
from bs4 import BeautifulSoup
import requests
from parse import *
```

```python
URL = "https://www.nettiauto.com/nayteikkuna-kaikki"
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:127.0) Gecko/20100101 Firefox/127.0"
}
nettiauto = requests.get(URL, headers=headers)
nettiauto.reason
```

```python
with open("nettiauto.html") as fp:
    soup = BeautifulSoup(fp)
soup.find_all("a")
```

```python
results = soup.find("div", class_="showcaseAdResult")
```

```python
results.find_all("li", class_="sms-section__window")[0]
```

```python
ads = results.find_all("li", class_="sms-section__window")

def return_ad_string(tag_str, class_str):
    elem = ad.find(tag_str, class_=class_str)
    if elem != None:
        return elem.string.strip()
    else:
        return "-"
    
for ad in ads:
    print(return_ad_string("div", "sms-section__vehicle-title"))
    print(return_ad_string("div", "sms-section__vehicle-price").split("\n")[0])
    for s in return_ad_string("span", "vehicle_tkm").split(","):
        print(s.strip())
    print(return_ad_string("span", "vehicle_tkm"))
    print()
```

```python
len(list(ads))
```

```python
test_price_str = "   11 000 \n €   "
test_price_str2 = "   1000 \n €   "

def format_price(price_str):
    # 11 000 € with white space -> 11000
    # 256 € with white space -> 256
    no_eur_sign = price_str.split("\n")[0].strip()
    return "".join(no_eur_sign.split(" "))

print(format_price(test_price_str2))
```

```python
test_str = "Renault Grand Scenic"

def split_title(title_str):
    split = title_str.strip().split(" ")
    return dict(brand=split[0], model=" ".join(split[1:]))

split_title(test_str)
```

```python
from parse import *

test_extra_str = "2, '21, 55 tkm"
test_extra_str2 = "'23, 32 tkm"

def split_extra(extra_str):
    split = extra_str.split(",")
    vol, year, tkm = None, None, None
    for s in split:
        if "tkm" in s:
            tkm = parse("{} tkm", s.strip())[0]
        elif "'" in s:
            year = parse("'{}", s.strip())[0]
        else:
            vol = s.strip()
    return dict(vol=vol, year=year, tkm=tkm)

split_extra(test_extra_str)
```

```python
ads = results.find_all("li", class_="sms-section__window")

cars = []

for ad in ads:
    car = dict()
    
    title_dict = split_title(return_ad_string("div", "sms-section__vehicle-title"))
    car["brand"] = title_dict["brand"]
    car["model"] = title_dict["model"]
    car["price"] = format_price(return_ad_string("div", "sms-section__vehicle-price"))

    extra_dict = split_extra(return_ad_string("span", "vehicle_tkm"))
    car["vol"] = extra_dict["vol"]
    car["year"] = extra_dict["year"]
    car["tkm"] = extra_dict["tkm"]

    cars.append(car)

len(cars)
```

```python
import pandas as pd

df = pd.DataFrame(cars)
df
```
